## 1. 划分数据集

```python
import pandas as pd
from sklearn.model_selection import train_test_split
# read from jsonlines file
data = None
# YOUR CODE HERE
data = pd.read_json('quotes_items.jl',lines=True)
# split into train and test
def split_data(data, train_size=0.8):
    train = None
    test = None
    # YOUR CODE HERE
    train, test = train_test_split(data, train_size=train_size,random_state=42)
    return train, test
train_set, test_set = split_data(data)
```

## 2. 选择/预处理特征

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
def get_features(data1, data2):
    features = None
    # YOUR CODE HERE
    vectorizer = TfidfVectorizer(max_features=1000, stop_words='english')
    train_features = vectorizer.fit_transform(data1['text'])
    test_features = vectorizer.transform(data2['text'])
    return train_features, test_features

train_set_features, test_set_features = get_features(train_set, test_set)

#将author转换为数字标签
encoder = LabelEncoder()
data_set_author = encoder.fit_transform(data['author'])
train_set_author = encoder.transform(train_set['author'])
test_set_author = encoder.transform(test_set['author'])
```

## 3. 训练模型

```python
from sklearn.neural_network import MLPClassifier
# declare a model
model = None
# YOUR CODE HERE
model = MLPClassifier(hidden_layer_sizes=(100,50), max_iter=1000, alpha=0.0001, solver='adam', verbose=10, random_state=21, tol=0.000000001)
# train the model
# YOUR CODE HERE
model.fit(train_set_features, train_set_author)
```

## 4. 测试模型

```python
# test the model with train set
# YOUR CODE HERE
train_predict = model.predict(train_set_features)
print(model.score(train_set_features, train_set_author))
# test the model with test set
# YOUR CODE HERE
test_predict = model.predict(test_set_features)
print(model.score(test_set_features, test_set_author))
test_set['predicted_author'] = encoder.inverse_transform(test_predict)
test_set
```
