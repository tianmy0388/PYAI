## 定义网络

```python
import torch
from torch import nn


batch_size = 16
num_epochs = 10
device = "cuda:0" if torch.cuda.is_available() else "cpu"

# YOUR CODE HERE to define the three CNNs
class Net_a(nn.Module):
    def __init__(self):
        super(Net_a, self).__init__()
        self.model = nn.Sequential(
            nn.Conv2d(in_channels=1, out_channels=10, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(10),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),

            nn.Conv2d(in_channels=10, out_channels=20, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(20),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),

            nn.Flatten(),
            nn.Linear(in_features=7*7*20, out_features=128),
            nn.ReLU(),
            nn.Linear(in_features=128, out_features=10),
            nn.Softmax(dim=1)
        )
    def forward(self, input):
        return self.model(input)

class Net_b(nn.Module):
    def __init__(self):
        super(Net_b, self).__init__()
        self.model = nn.Sequential(
            nn.Conv2d(in_channels=1, out_channels=10, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),

            nn.Conv2d(in_channels=10, out_channels=20, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),

            nn.Flatten(),
            nn.Linear(in_features=7*7*20, out_features=128),
            nn.ReLU(),
            nn.Linear(in_features=128, out_features=10),
            nn.Softmax(dim=1)
        )
    def forward(self, input):
        return self.model(input)

class Net_c(nn.Module):
    def __init__(self):
        super(Net_c, self).__init__()
        self.model = nn.Sequential(
            nn.Conv2d(in_channels=1, out_channels=16, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),

            nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),

            nn.Flatten(),
            nn.Linear(in_features=7*7*32, out_features=256),
            nn.ReLU(),
            nn.Linear(in_features=256, out_features=10),
            nn.Softmax(dim=1)
        )
    def forward(self, input):
        return self.model(input)

net_a = Net_a().to(device)
net_b = Net_b().to(device)
net_c = Net_c().to(device)
```

## 定义数据与预处理

```python
from torchvision import transforms


# define image transform
transform = transforms.Compose(
    [
        transforms.ToTensor(),
    ]
)
```

```python
from torchvision.datasets import MNIST
from torch.utils.data import DataLoader


# load data
train_data = MNIST(root='./data', train=True, download=True, transform=transform)
test_data = MNIST(root='./data', train=False, download=True, transform=transform)
train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True, num_workers=4)
test_loader = DataLoader(test_data, batch_size=batch_size, shuffle=False, num_workers=4)
```

## 定义优化器与目标函数

```python
from torch import optim


def get_optimizer(net):
    return optim.SGD(net.parameters(), lr=0.001, momentum=0.9, weight_decay=0.0001)

# setup the loss function
loss_fn = nn.CrossEntropyLoss()
```

## 定义训练过程

```python
from tqdm import tqdm
import torch


def train(train_loader, net, optimizer, loss_fn):
    # set model to train mode
    net.train()
    corrects = 0
    for inputs, labels in tqdm(train_loader, leave=False):
        # YOUR CODE BELOW to implement one training step
        # 0. move inputs and labels to device
        inputs = inputs.to(device)
        labels = labels.to(device)
        # 1. zero the parameter gradients
        optimizer.zero_grad()
        # 2. forward
        outputs = net(inputs)
        # 3. loss
        loss = loss_fn(outputs, labels)
        # 4. backward and optimize
        loss.backward()
        optimizer.step()
        # YOUR CODE ABOVE

        preds = outputs.argmax(1).detach()
        corrects += (preds==labels.data).sum()
    return loss, corrects / len(train_loader.dataset)

def test(test_loader, net):
    # set model to eval mode
    net.eval()
    corrects = 0
    with torch.no_grad():
        for inputs, labels in tqdm(test_loader, leave=False):
            # YOUR CODE BELOW to implement one testing step
            # 0. move inputs and labels to device
            inputs = inputs.to(device)
            labels = labels.to(device)
            # 1. forward
            outputs = net(inputs)
            # YOUR CODE ABOVE

            preds = outputs.argmax(1).detach()
            corrects += (preds==labels.data).sum()
    return corrects / len(test_loader.dataset)
```

## 执行训练、测试和比较

```python
for net in [net_a, net_b, net_c]:
    # YOUR CODE HERE to train, evaluate, and compare the three models
    optimizer = get_optimizer(net)
    train(train_loader, net, optimizer, loss_fn)
    print(test(test_loader, net).item())
```
